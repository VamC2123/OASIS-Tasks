```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import r2_score  
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
#Performing EDA
#Handling Binary Categorical Variables (Yes/No) Category
```

```python
categorical_col =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[categorical_col]
```

```python
def binary_map(categorical_col):
    return categorical_col.map({'yes': 1, 'no': 0})
```

```python
df[categorical_col] = df[categorical_col].apply(binary_map)

df[categorical_col]
```

```python
df.head()
```

```python
#Handling Categorical data with Dummy Variable
```

```python
dummy_col = pd.get_dummies(df['furnishingstatus'])
dummy_col.head()
```

```python
dummy_col = pd.get_dummies(df['furnishingstatus'], drop_first=True)
dummy_col.head()
```

```python
df = pd.concat([df, dummy_col], axis=1)
df.head()
```

```python
df.drop(['furnishingstatus'], axis=1, inplace=True)
df.head()
```

```python
#Splitting data into Training and Testing data
```

```python
df.columns
```

```python
np.random.seed(0)

df1_train, df1_test = train_test_split(df, train_size=0.7, test_size=0.3, random_state=100)
```

```python
df1_train.head()
```

```python
df1_train.shape
```

```python
df1_test.head()
```

```python
df1_test.shape
```

```python
#Scaling Training Data: MinMaxScaler
```

```python
scaler = MinMaxScaler()
col_to_scale = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']
df1_train[col_to_scale] = scaler.fit_transform(df1_train[col_to_scale])
```

```python
#Training the model
```

```python
y_train = df1_train.pop('price')
x_train = df1_train
linear_regression = LinearRegression()
```

```python
linear_regression.fit(x_train, y_train)
```

```python
score = linear_regression.score(x_train, y_train)
```

```python
#Scaling Test Data: MinMaxScaler
```

```python
col_to_scale = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']
```

```python
df1_test[col_to_scale] = scaler.fit_transform(df1_test[col_to_scale])
```

```python
#Testing our model
```

```python
y_test = df1_test.pop('price')
x_test = df1_test
```

```python
prediction = linear_regression.predict(x_test)
```

```python
#Checking R squared value
```

```python
r2 = r2_score(y_test, prediction)
```

```python
mse = mean_squared_error(y_test, prediction)
print("Mean Squared Error:", mse)
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5) 
knn_model.fit(x_train, y_train)
knn_y_pred = knn_model.predict(x_test)
knn_mse = mean_squared_error(y_test, knn_y_pred)
knn_r2 = r2_score(y_test, knn_y_pred)
```

```python
print(" Mean Squared Error:", knn_mse)
print(" R-squared:", knn_r2)
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(x_train, y_train)
dt_y_pred = dt_model.predict(x_test)
```

```python
dt_mse = mean_squared_error(y_test, dt_y_pred)
dt_r2 = r2_score(y_test, dt_y_pred)
```

```python
print(" Mean Squared Error:", dt_mse)
print(" R-squared:", dt_r2)
```

```python
algorithms = ['Linear Regression', 'KNN', 'Decision Tree']
r2_scores = [r2, knn_r2, dt_r2]

plt.figure(figsize=(10, 6))
plt.bar(algorithms, r2_scores, color=['blue', 'green', 'red'])
plt.title('R-squared Comparison')
plt.xlabel('Algorithm')
plt.ylabel('R-squared')
plt.ylim(0, 1)  
plt.show()
```
